```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/bi.csv' , encoding="latin1")
df.head()
```

```python
for col in df.columns:
  print(col   , df[col].unique())
  print("-"*50)
```

```python
df['gender'] = df['gender'].replace(('F' , 'female'),'Female')
df['gender'] = df['gender'].replace(('M' , 'male'),'Male')
```

```python
df['FullName'] = df['fNAME'] + ' ' + df['lNAME']
```

```python
df = df.drop(columns=['fNAME','lNAME'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Python'] = df['Python'].fillna(df['Python'].mean())

```

```python
df['GPA'] = (df['DB'] + df['Python'] )/ 2
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error

le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
df['country'] = le.fit_transform(df['country'])
df['residence'] = le.fit_transform(df['residence'])
df['prevEducation'] = le.fit_transform(df['prevEducation'])
df.head()
```

```python
df_ml = df.drop(columns=['FullName'])
#df_ml.head()
```

```python
sv = SVR(kernel='rbf', C=10, epsilon=0.1, gamma='scale')
X = df_ml.drop(columns=['GPA'])
y = df_ml['GPA']
X_train , X_test , y_train , y_test = train_test_split(X , y , random_state=42 , test_size=0.2)
y_pred = sv.fit(X_train , y_train).predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("R2 Score:", r2)
```
