```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('/content/house-prices.csv')
data.head(7)
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, axes = plt.subplots(3, 1, figsize=(15, 15))
sns.histplot(data['Price'], kde=True, color='skyblue' ,ax=axes[0])
axes[0].set_title("Distribution of Price")

plt.xlabel("Price")
plt.ylabel("Frequency")

# Distribution of SqFt

sns.histplot(data['SqFt'], kde=True, color='green',ax=axes[1])
axes[1].set_title("Distribution of SqFt")
plt.xlabel("SqFt")
plt.ylabel("Frequency")

# Distribution of Bedrooms

sns.histplot(data['Bedrooms'], kde=True, color='purple',ax=axes[2])
axes[2].set_title("Distribution of Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Frequency")
```

```python
sns.countplot(x='Brick', data=data, palette="coolwarm") # if yes or no
plt.title("Count of Brick Houses")
plt.xlabel("Brick")
plt.ylabel("Count")
```

```python
sns.countplot(x='Neighborhood', data=data, palette="viridis") # depend (west, east, north)
plt.title("Houses per Neighborhood")
plt.xlabel("Neighborhood")
plt.ylabel("Count")
```

```python
sns.scatterplot(x='SqFt', y='Price', data=data, hue='Neighborhood', palette="deep")
plt.title("Price vs. SqFt")
plt.xlabel("SqFt")
plt.ylabel("Price")
```

```python
sns.scatterplot(x='Bedrooms', y='Price', data=data, hue='Brick', palette="coolwarm")
plt.title("Price vs. Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
```

```python
numeric_columns = data.select_dtypes(include=['number'])# Select only numeric columns for correlation
new_correlation = numeric_columns.corr()
sns.heatmap(new_correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
```

```python
mean_price = np.mean(data['Price'])
median_price = np.median(data['Price'])
std_price = np.std(data['Price'])
var_price = np.var(data['Price'])
min_price = np.min(data['Price'])
max_price = np.max(data['Price'])
print(f'Mean Price: {mean_price}')
print(f'Median Price: {median_price}')
print(f'Standard Deviation of Price: {std_price}')
print(f'Variance of Price: {var_price}')
print(f'Minimum Price: {min_price}')
print(f'Maximum Price: {max_price}')
```
